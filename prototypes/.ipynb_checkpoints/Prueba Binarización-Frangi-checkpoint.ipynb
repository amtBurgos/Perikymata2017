```python
#Imports
from skimage import io
from skimage.color import rgb2grey
from skimage.morphology import skeletonize
from skimage.color import rgb2lab,gray2rgb
from skimage import exposure
from skimage.restoration import denoise_tv_chambolle
from _frangi import frangi, hessian
from skimage.filters import threshold_li
from skimage import morphology
import warnings as war
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from os import walk
from os import path
from scipy.ndimage.filters import prewitt
import csv
from scipy import misc
```

```python
# Importar imágenes
img = io.imread('Optimize\\Image.bmp')

#Carpeta donde guardarla
folder = ('Binarizacion-Frangi-Images-Test\\')
```

```python
#Binarización
def sinFrangi(img,removeSmall=5):
    result = None
    # disable warnings
    with war.catch_warnings():
        war.simplefilter("ignore")
        img = rgb2grey(img)
        imgAdapted = exposure.equalize_adapthist(img, clip_limit=0.91,nbins =100)
        imgAdaptedDenoise = denoise_tv_chambolle(imgAdapted, weight=0.1)
        pwt=prewitt(imgAdaptedDenoise)
        threshF = threshold_li(pwt)
        thresholdedF = pwt >= threshF
        skF=skeletonize(thresholdedF)
        no_smallF = morphology.remove_small_objects(skF, min_size=removeSmall,connectivity =100)
    return no_smallF
```

```python
#Frangi
def todoProceso(img,removeSmall=5):
    with war.catch_warnings():
        war.simplefilter("ignore")
        img = rgb2grey(img)
        imgAdapted = exposure.equalize_adapthist(img, clip_limit=0.91,nbins =100)
        imgAdaptedDenoise = denoise_tv_chambolle(imgAdapted, weight=0.1)
        pwt=prewitt(imgAdaptedDenoise)
        frangiImg = frangi(pwt)
        threshF = threshold_li(frangiImg)
        thresholdedF = frangiImg >= threshF
        skF=skeletonize(thresholdedF)
        no_smallF = morphology.remove_small_objects(skF, min_size=removeSmall,connectivity =100)
    return no_smallF
```

```python
def sinEqualizacion(img, removeSmall=5):
    with war.catch_warnings():
        war.simplefilter("ignore")
        img = rgb2grey(img)
        pwt=prewitt(img)
        frangiImg = frangi(pwt)
        threshF = threshold_li(frangiImg)
        thresholdedF = frangiImg >= threshF
        skF=skeletonize(thresholdedF)
        no_smallF = morphology.remove_small_objects(skF, min_size=removeSmall,connectivity =100)
    return no_smallF
```

```python
binarized = sinFrangi(img)
allProcess = todoProceso(img)
```

```python
misc.toimage(binarized, cmin=False, cmax=True).save(folder+'binarized.jpg')
misc.toimage(allProcess, cmin=False, cmax=True).save(folder+'allProcess.jpg')
```

```python
plt.figure();
plt.title("Procesado SIN frangi");
plt.imshow(binarized,cmap='gray')
```

```python
plt.figure();
plt.title("Procesado CON frangi");
plt.imshow(allProcess,cmap='gray')
```

```python
whitoutEqualization = sinEqualizacion(img)
```

```python
misc.toimage(whitoutEqualization, cmin=False, cmax=True).save(folder+'whitoutEqualization.jpg')
```

```python
plt.figure()
plt.title('Sin equalización')
plt.imshow(whitoutEqualization, cmap='gray')
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

#Verlo interactivamente
def interactivo(image, tipo, removeSmall):
    result = None  
    if tipo=='Sin Frangi':
        result = sinFrangi(image, removeSmall)
    elif tipo=='Sin Equializacion':
        result = sinEqualizacion(image, removeSmall)
    elif tipo =='Todo':
        result = todoProceso(image, removeSmall)        
    plt.figure();
    plt.imshow(result, cmap='gray');
```
