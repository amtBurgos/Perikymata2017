```python
#Imports
from skimage import io
from skimage.color import rgb2grey
from skimage.morphology import skeletonize
from skimage.color import rgb2lab,gray2rgb
from skimage import exposure
from skimage.restoration import denoise_tv_chambolle
from _frangi import frangi, hessian
from skimage.filters import threshold_li
from skimage import morphology
import warnings as war
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from os import walk
from os import path
from scipy.ndimage.filters import prewitt
import csv
```

```python
# Importar imágenes
img = io.imread('Optimize\\Image.bmp')
```

```python
#Binarización
def sinFrangi(img):
    result = None
    # disable warnings
    with war.catch_warnings():
        war.simplefilter("ignore")
        img = rgb2grey(img)
        imgAdapted = exposure.equalize_adapthist(img, clip_limit=0.91,nbins =100)
        imgAdaptedDenoise = denoise_tv_chambolle(imgAdapted, weight=0.1)
        pwt=prewitt(imgAdaptedDenoise)
        threshF = threshold_li(pwt)
        thresholdedF = pwt >= threshF
        skF=skeletonize(thresholdedF)
        no_smallF = morphology.remove_small_objects(skF, min_size=5,connectivity =100)
    return no_smallF
```

```python
#Frangi
def todoProceso(img):
    with war.catch_warnings():
        war.simplefilter("ignore")
        img = rgb2grey(img)
        imgAdapted = exposure.equalize_adapthist(img, clip_limit=0.91,nbins =100)
        imgAdaptedDenoise = denoise_tv_chambolle(imgAdapted, weight=0.1)
        pwt=prewitt(imgAdaptedDenoise)
        frangiImg = frangi(pwt)
        threshF = threshold_li(frangiImg)
        thresholdedF = frangiImg >= threshF
        skF=skeletonize(thresholdedF)
        no_smallF = morphology.remove_small_objects(skF, min_size=5,connectivity =100)
    return no_smallF
```

```python
def sinEqualizacion(img):
    with war.catch_warnings():
        war.simplefilter("ignore")
        img = rgb2grey(img)
        pwt=prewitt(img)
        frangiImg = frangi(pwt)
        threshF = threshold_li(frangiImg)
        thresholdedF = frangiImg >= threshF
        skF=skeletonize(thresholdedF)
        no_smallF = morphology.remove_small_objects(skF, min_size=5,connectivity =100)
    return no_smallF
```

```python
binarized = sinFrangi(img)
allProcess = todoProceso(img)
```

```python
plt.figure();
f, axes = plt.subplots(1,2,figsize=(15, 15));
for axe in axes:
    axe.axis('off');

axes[0].set_title("Procesado SIN frangi");
axes[1].set_title("Procesado CON frangi");

axes[0].imshow(binarized, cmap='gray');
axes[1].imshow(allProcess, cmap='gray');
```

```python
whitoutEqualization = sinEqualizacion(img)
```

```python
plt.figure()
plt.title('Sin equalización')
plt.imshow(whitoutEqualization, cmap='gray')
```
