```python
imgSk2, lines2 = kirsch.kirschProcessing(imgPrepared, kernelId=2)
kirsch.showImage(imgSk2)
```

```python
fig = plt.figure(figsize=(25, 16))
plt.imshow(imgSk2, cmap=plt.cm.gray)

plt.title('Probabilistic Hough')
for line in lines+lines2:
    p0, p1 = line
    plt.plot((p0[0], p1[0]), (p0[1], p1[1]),'r',linewidth=1)
plt.show()
```
